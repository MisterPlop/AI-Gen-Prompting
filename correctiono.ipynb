```python
# Importer les librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.model_selection import train_test_split

import xgboost as xgb

from sklearn.metrics import mean_squared_error, r2_score

from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline

import pickle

import warnings
warnings.filterwarnings("ignore")
```



---






Generate Python code to load a CSV file named 'avocado.csv' into a pandas DataFrame.



```python
# Charger le dataset
data = pd.read_csv('/content/drive/MyDrive/EPSI/avocado.csv')
```



---




Remove unnecessary columns: 'Unnamed: 0', 'Total Volume', 'Total Bags'.


```python
# Supprimer les colonnes inutiles ('Unnamed: 0','Total Volume', 'Total Bags')
df = data.drop(['Unnamed: 0','Total Volume', 'Total Bags'], axis=1)
```



---



Rename the following columns: "4046" to "Quality1", "4225" to "Quality2", "4770" to "Quality3".


```python
# Renommer les colonnes "4046" : "Quality1", "4225" : "Quality2","4770" : "Quality3"
df = df.rename(index=str, columns={"4046" : "Quality1", "4225" : "Quality2","4770" : "Quality3" })
```



---



Convert the 'Date' column to pandas datetime format.


```python
# Encoder les dates en to_datetime
df["Date"] = pd.to_datetime(df["Date"])
```

Display the count of duplicates.



---



Generate Python code to define two lists of feature names:
1. 'num_features' for the numeric columns: ['Quality1', 'Quality2', 'Quality3', 'Small Bags', 'Large Bags', 'XLarge Bags', 'year']
2. 'cat_features' for the categorical columns: ['type', 'region']

```python
# Définir les colonnes numériques et catégoriques
num_features = ['Quality1', 'Quality2', 'Quality3', 'Small Bags', 'Large Bags', 'XLarge Bags', 'year']
cat_features = ['type', 'region']
```



---



Generate Python code to create two transformers for feature preprocessing:
1. A numerical transformer using StandardScaler.
2. A categorical transformer using OneHotEncoder with `drop='first'` and `handle_unknown='ignore'`.

```python
# Pipeline pour colonnes numériques
num_transformer = StandardScaler()
```

```python
# Pipeline pour colonnes catégoriques
cat_transformer = OneHotEncoder(drop='first', handle_unknown='ignore')
```



---



Generate Python code to create a ColumnTransformer that applies:
1. The numerical transformer to the 'num_features' list.
2. The categorical transformer to the 'cat_features' list.

```python
# Combiner les transformations dans un ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)
    ]
)
```





---






Generate Python code to create a full machine learning pipeline that includes:
1. Preprocessing steps using the ColumnTransformer created earlier.
2. A machine learning model, using XGBRegressor from xgboost.


```python
# Créer un pipeline global avec prétraitement + modèle
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', xgb.XGBRegressor())
])
```



---



Generate Python code to split the dataset into features (X) and target (y), with 'AveragePrice' as the target variable.
Then split the data into training and test sets using train_test_split, with a test size of 20% and a random state of 42.

```python
# Diviser les données en ensembles d'entraînement et de test
X = df.drop(columns='AveragePrice')
y = df['AveragePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```



---



Generate Python code to train the machine learning pipeline using the training data (X_train, y_train).


```python
# Entraîner le pipeline
pipeline.fit(X_train, y_train)
```



---



Generate Python code to evaluate the model using the test data (X_test, y_test):
1. Make predictions using the trained model.
2. Calculate the RMSE (Root Mean Squared Error) and R2 (R-squared) score.
3. Print both metrics.


```python
# Évaluer le pipeline
pred = pipeline.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, pred))
r2 = r2_score(y_test, pred)
```

```python
# Afficher les métriques
print(f'RMSE --> {rmse}')
print(f'R2 --> {r2}')
```



---



Generate Python code to save the entire pipeline, including preprocessing and the trained model, into a file named 'pipeline.pkl' using pickle.


```python
# Sauvegarder le pipeline entier
with open('pipeline.pkl', 'wb') as file:
    pickle.dump(pipeline, file)
```
